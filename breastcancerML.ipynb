MACHINE LEARNING

While the size of our features is quiltely large, We are going to use the PCA (principal components analysis) in order to decrease the dimensionality of the data. 

```python
from sklearn.preprocessing import StandardScaler, normalize
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
from factor_analyzer.factor_analyzer import calculate_kmo
from sklearn.decomposition import PCA
```

```python
x=data.iloc[:,2:]
y=data.iloc[:,1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_sc=scaler.fit_transform(x)
from sklearn.preprocessing import normalize
x_n=normalize(x_sc)
x=pd.DataFrame(x_n,columns=x.columns)
x
```

```python
chi_square_value,p_value=calculate_bartlett_sphericity(x)
chi_square_value, p_value 
#p_value=0 strong correlation between features(no need to use some of them -> use all the features.
```

```python
kmo_all,kmo_model=calculate_kmo(x)
kmo_model
```

The p_value<0.05 &  the kmo_model>0.6 . So the PCA is applicable on our data.

```python
fa = FactorAnalyzer(rotation='varimax')
#FactorAnalyzer: This is a class from the factor_analyzer library used to perform factor analysis, a technique for identifying underlying relationships between variables.
fa.fit(x)
# Check Eigenvalues
ev, v = fa.get_eigenvalues()
# Create scree plot using matplotlib
plt.scatter(range(1,x.shape[1]+1),ev)
plt.plot(range(1,x.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

The number of the factors has to be choosen as the first number upon the axis of Eigenvalue=1. With reference to this figure , n_factors=6.

```python
fa = FactorAnalyzer(n_factors=6,rotation='varimax')
fa.fit(x)
factor_df=pd.DataFrame(fa.loadings_,index=x.columns)
print(factor_df)
```

```python
print(pd.DataFrame(fa.get_factor_variance(),index=['Variance','Proportional Var','Cumulative Var']))
```

We are going to use the 5 components of the PCA, while it's cumulative variance is equal to 0.84, which is a very good value.

```python
pca = PCA(n_components=6)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns=['p1', 'p2', 'p3','p4','p5','p6'])
principalDf
```

```python
new_data=pd.concat([principalDf,y],axis=1)
new_data
```
